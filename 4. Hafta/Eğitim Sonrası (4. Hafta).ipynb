````python
# 2.Problemin a ÅÄ±kkÄ±nÄ±n Ã‡Ã¶zÃ¼mÃ¼nÃ¼ Buraya YazÄ±nÄ±z
numbers = [5, 10, 10, 15, 15, 15, 20, 20, 20, 20]
unique = []
for number in numbers:
    if number in unique:
        continue
    else:
        unique.append(number)
print(unique)
print(numbers)
````
