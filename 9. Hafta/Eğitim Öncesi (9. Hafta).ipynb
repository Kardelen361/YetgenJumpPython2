# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def vehicle_info(self):
        print(self.make, self.model, self.year)


class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        print("The seating capacity of this bus is:", capacity)


my_bus = Bus("Mercedes", "Travego", 2022)
my_bus.vehicle_info()
my_bus.seating_capacity()
````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
# 2.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, max_speed, mileage):
        self.max_speed = max_speed
        self.mileage = mileage
    
class Bus(Vehicle):
    def __init__(self, max_speed, mileage, seating_capacity=50):
        super().__init__(max_speed, mileage)
        self.seating_capacity = seating_capacity

class School_bus(Bus):
    def __init__(self, max_speed, mileage, seating_capacity=50):
        super().__init__(max_speed, mileage, seating_capacity)

sb = School_bus(60, 10000)
print(isinstance(sb, Vehicle))
````
